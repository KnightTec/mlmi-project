```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib

print(nib.load("data/brats_2021/TrainingData/BraTS2021_00000/BraTS2021_00000_seg.nii.gz").get_fdata().shape)
imgs = [nib.load(f"data/brats_2021/TrainingData/BraTS2021_00000/BraTS2021_00000_{m}.nii.gz").get_fdata().astype(np.float32)[:, :, 75] for m in ["flair", "t1", "t1ce", "t2"]]
lbl = nib.load("data/brats_2021/TrainingData/BraTS2021_00000/BraTS2021_00000_seg.nii.gz").get_fdata().astype(np.uint8)[:, :, 75]
fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(15, 15))
for i, img in enumerate(imgs):
    ax[i].imshow(img.T, cmap='gray')
    ax[i].axis('off')
ax[-1].imshow(lbl.T, vmin=0, vmax=4)
ax[-1].axis('off')
plt.tight_layout()            
plt.show()
```
