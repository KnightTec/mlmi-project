```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'T1':4116, 'T2':4051, 'T2*':4051, 'FLAIR':1381, "TOF MRA": 896}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 6))
 
# creating the bar plot
plt.bar(courses, values, width = 0.6)

i = 0
for key, value in data.items():
    plt.text(i, value + 50, value, ha = 'center')
    i += 1
 
plt.xlabel("MR Modalites")
plt.ylabel("Scan Count")
plt.title("Number of sMRI scans in OASIS-3 w/ duplicate modalities")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("csv/oasis/oasis_3_mulitmodal_full.csv", sep=";")

data = {'CDR = 0': (df["cdr"] == 0).sum(), 'CDR > 0':(df["cdr"] > 0).sum()}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 6))
 
# creating the bar plot
plt.bar(courses, values, width = 0.6)

i = 0
for key, value in data.items():
    plt.text(i, value + 50, value, ha = 'center')
    i += 1
 
plt.xlabel("Clinical Dementia Rating")
plt.ylabel("Count")
plt.title("Alzheimer's Disease Labels Counts")
plt.show()
```
