```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'T1':4116, 'T2':4051, 'T2*':4051, 'FLAIR':1381, "TOF MRA": 896}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 6))
 
# creating the bar plot
plt.bar(courses, values, width = 0.6)

i = 0
for key, value in data.items():
    plt.text(i, value + 50, value, ha = 'center')
    i += 1
 
plt.xlabel("MR Modalites")
plt.ylabel("Scan Count")
plt.title("Number of sMRI scans in OASIS-3 w/ duplicate modalities")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("csv/oasis/oasis_3_mulitmodal_full.csv", sep=";")
df.fillna('', inplace=True)

t1_count = df['MR T1w'].apply(lambda x: len(x.strip()) > 0).sum()
t2_count = df['MR T2w'].apply(lambda x: len(x.strip()) > 0).sum()
t2_star_count = df['MR T2*'].apply(lambda x: len(x.strip()) > 0).sum()
flair_count = df['MR FLAIR'].apply(lambda x: len(x.strip()) > 0).sum()
tof_mra_count = df['MR TOF-MRA'].apply(lambda x: len(x.strip()) > 0).sum()


# creating the dataset
data = {'T1':t1_count, 'T2':t2_count, 'T2*':t2_star_count, 'FLAIR':flair_count, "TOF MRA": tof_mra_count}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 6))
 
# creating the bar plot
plt.bar(courses, values, width = 0.6)

i = 0
for key, value in data.items():
    plt.text(i, value + 50, value, ha = 'center')
    i += 1
 
plt.xlabel("MR Modalites")
plt.ylabel("Scan Count")
plt.title("Number of sMRI scans in OASIS-3 w/o duplicate modalities")
plt.show()
```
